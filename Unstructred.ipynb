```python
from unstructured.partition.pdf import partition_pdf
import time

def extract_pdf_elements(filename: str, strategy: str = "hi_res"):
    # Start timing before extraction begins
    start_time = time.time()

    # Extract elements from the PDF using the specified strategy and table structure inference
    elements = partition_pdf(
        filename=filename,
        strategy=strategy,
        infer_table_structure=True
    )

    # Capture the extraction time
    extraction_time = time.time() - start_time
    print(f"Total extraction time: {extraction_time:.2f} seconds")

    # Separate elements into categories based on their .category attribute
    text_elements = [el for el in elements if el.category == "Text"]
    table_elements = [el for el in elements if el.category == "Table"]
    image_elements = [el for el in elements if el.category == "Image"]

    print(f"Found {len(text_elements)} text element(s), {len(table_elements)} table(s), and {len(image_elements)} image(s).\n")

    # Process table elements
    print("--- Tables ---")
    for idx, table in enumerate(table_elements, start=1):
        print(f"\nTable {idx} Text:")
        print(table.text)
        # Print HTML representation of the table if available
        if hasattr(table, "metadata") and hasattr(table.metadata, "text_as_html"):
            print("Table Structure (HTML):")
            print(table.metadata.text_as_html)
        else:
            print("No HTML metadata available for this table.")

    # Process text elements
    print("\n--- Text ---")
    for idx, text in enumerate(text_elements, start=1):
        print(f"\nText Element {idx}:")
        print(text.text)

    # Process image elements
    print("\n--- Images ---")
    for idx, image in enumerate(image_elements, start=1):
        print(f"\nImage {idx}:")
        # If binary data is available, display its size (you can extend this to save or display the image)
        if hasattr(image, "binary_data") and image.binary_data:
            print(f"Image binary data size: {len(image.binary_data)} bytes")
        else:
            print("No binary data available for this image.")

    return {
        "extraction_time": extraction_time,
        "text_elements": text_elements,
        "table_elements": table_elements,
        "image_elements": image_elements,
    }

# Example usage:
filename = "VERIZON USER GUIDE.pdf"  # Replace with your document's path
result = extract_pdf_elements(filename)
```
