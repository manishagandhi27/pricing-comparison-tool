```python
from unstructured.partition.pdf import partition_pdf
import time

def extract_pdf_elements(filename: str, strategy: str = "hi_res"):
    # Start timing before extraction begins
    start_time = time.time()

    # Extract elements from the PDF using the specified strategy and table structure inference
    elements = partition_pdf(
        filename=filename,
        strategy=strategy,
        infer_table_structure=True
    )

    # Capture the extraction time
    extraction_time = time.time() - start_time
    print(f"Total extraction time: {extraction_time:.2f} seconds")

    # Separate elements into categories based on their .category attribute
    text_elements = [el for el in elements if el.category == "Text"]
    table_elements = [el for el in elements if el.category == "Table"]
    image_elements = [el for el in elements if el.category == "Image"]

    print(f"Found {len(text_elements)} text element(s), {len(table_elements)} table(s), and {len(image_elements)} image(s).\n")

    # Process table elements
    print("--- Tables ---")
    for idx, table in enumerate(table_elements, start=1):
        print(f"\nTable {idx} Text:")
        print(table.text)
        # Print HTML representation of the table if available
        if hasattr(table, "metadata") and hasattr(table.metadata, "text_as_html"):
            print("Table Structure (HTML):")
            print(table.metadata.text_as_html)
        else:
            print("No HTML metadata available for this table.")

    # Process text elements
    print("\n--- Text ---")
    for idx, text in enumerate(text_elements, start=1):
        print(f"\nText Element {idx}:")
        print(text.text)

    # Process image elements
    print("\n--- Images ---")
    for idx, image in enumerate(image_elements, start=1):
        print(f"\nImage {idx}:")
        # If binary data is available, display its size (you can extend this to save or display the image)
        if hasattr(image, "binary_data") and image.binary_data:
            print(f"Image binary data size: {len(image.binary_data)} bytes")
        else:
            print("No binary data available for this image.")

    return {
        "extraction_time": extraction_time,
        "text_elements": text_elements,
        "table_elements": table_elements,
        "image_elements": image_elements,
    }

# Example usage:
filename = "VERIZON USER GUIDE.pdf"  # Replace with your document's path
result = extract_pdf_elements(filename)
```

```python
from copy import deepcopy
from sentence_transformers import SentenceTransformer, util
import time

# Initialize the model for text similarity.
model = SentenceTransformer('all-MiniLM-L6-v2')

# --- Utility to ensure we have a dictionary ---
def to_dict(table_item):
    if isinstance(table_item, dict):
        return table_item
    elif hasattr(table_item, "dict"):
        return table_item.dict()
    else:
        return table_item

def bbox_match(bbox1, bbox2, threshold=10.0):
    left_diff = abs(bbox1['l'] - bbox2['l'])
    right_diff = abs(bbox1['r'] - bbox2['r'])
    top_diff = abs(bbox1['t'] - bbox2['t'])
    return (left_diff <= threshold and right_diff <= threshold and top_diff <= threshold)

def merge_bbox(bbox1, bbox2):
    return {
        'l': min(bbox1['l'], bbox2['l']),
        'r': max(bbox1['r'], bbox2['r']),
        'b': min(bbox1['b'], bbox2['b']),
        't': max(bbox1['t'], bbox2['t']),
        'coord_origin': bbox1.get('coord_origin', 'BOTTOMLEFT')
    }

def is_at_page_end(bbox, page_height, fraction=0.3):
    # Table is considered at the bottom if its bottom coordinate is less than 30% of page height.
    return bbox['b'] <= fraction * page_height

def is_at_page_top(bbox, page_height, fraction=0.7):
    # Table is considered at the top if its top coordinate is at least 70% of page height.
    return bbox['t'] >= fraction * page_height

def get_col_count(table):
    grid = table.get('data', {}).get('grid', [])
    return len(grid[0]) if grid and grid[0] else 0

def extract_first_row_text(table):
    grid = table.get('data', {}).get('grid', [])
    if grid and grid[0]:
        return " | ".join(cell.get('text', '').strip() for cell in grid[0])
    return ""

def compute_text_similarity(text1, text2):
    if not text1 or not text2:
        return 0.0
    emb1 = model.encode(text1, convert_to_tensor=True)
    emb2 = model.encode(text2, convert_to_tensor=True)
    return util.pytorch_cos_sim(emb1, emb2).item()

def merge_tables_list(tables, bbox_threshold=10.0, header_sim_threshold=0.7, page_height=792):
    """
    Merge multi-page table fragments using positional, structural, and header content cues.

    For each candidate table:
      1. It must be on the immediate next page.
      2. The current table's bbox must be at the bottom (bottom coordinate <= 30% of page height)
         and the candidate's bbox must be at the top (top coordinate >= 70% of page height).
      3. Their bounding boxes (l, r, t) must match within a threshold.
      4. They must have the same number of columns.
      5. Then compare header text; if very similar, skip the candidate's header row.

    A new key "merged_ids" is added to record which self_ref IDs were merged.
    """
    tables = [to_dict(t) for t in tables]
    merged_tables = []
    i = 0
    while i < len(tables):
        current_table = deepcopy(tables[i])
        merge_ids = [current_table.get('self_ref', f"Table_{i}")]
        current_page = current_table['prov'][0]['page_no']
        current_bbox = current_table['prov'][0]['bbox']
        current_cols = get_col_count(current_table)
        merged_grid = current_table.get('data', {}).get('grid', [])[:]

        print(f"\nStarting merge with {merge_ids[0]} on page {current_page}")

        j = i + 1
        while j < len(tables):
            candidate = tables[j]
            candidate_self_ref = candidate.get('self_ref', f"Table_{j}")
            candidate_page = candidate['prov'][0]['page_no']

            # Only merge if the candidate is on the immediate next page.
            if candidate_page != current_page + 1:
                break

            # Condition 2: Positional cues using relative thresholds.
            if not (is_at_page_end(current_bbox, page_height, fraction=0.3) and
                    is_at_page_top(candidate_bbox, page_height, fraction=0.7)):
                break

            # Condition 3: Bounding box similarity.
            if not bbox_match(current_bbox, candidate_bbox, threshold=bbox_threshold):
                break

            # Condition 4: Structural consistency (same number of columns).
            candidate_cols = get_col_count(candidate)
            if candidate_cols != current_cols:
                break

            # Condition 5: Check header similarity.
            current_header_text = extract_first_row_text(current_table)
            candidate_first_text = extract_first_row_text(candidate)
            header_sim = compute_text_similarity(current_header_text, candidate_first_text)
            if header_sim >= header_sim_threshold:
                print(f"--> Merging {candidate_self_ref} (page {candidate_page}): duplicate header detected (sim={header_sim:.2f}), skipping candidate's header row.")
                candidate_grid = candidate.get('data', {}).get('grid', [])
                merge_rows = candidate_grid[1:] if len(candidate_grid) > 1 else []
            else:
                print(f"--> Merging {candidate_self_ref} (page {candidate_page}): header not duplicate (sim={header_sim:.2f}), merging all rows.")
                merge_rows = candidate.get('data', {}).get('grid', [])

            merge_ids.append(candidate_self_ref)
            merged_grid.extend(merge_rows)

            # Update bounding box and page info.
            current_bbox = merge_bbox(current_bbox, candidate_bbox)
            current_table['prov'][0]['bbox'] = current_bbox
            current_page = candidate_page

            j += 1

        current_table['data']['grid'] = merged_grid
        current_table['merged_ids'] = merge_ids
        merged_tables.append(current_table)
        i = j

    return merged_tables


# --- Example Usage ---
if __name__ == "__main__":
    # The following imports and document conversion are specific to your setup.
    # They are provided here as an example.
    from docling.document_converter import DocumentConverter
    from docling.backend.pypdfium2_backend import PyPdfiumDocumentBackend
    from docling.datamodel.base_models import InputFormat
    from docling.datamodel.pipeline_options import (
        AcceleratorDevice,
        AcceleratorOptions,
        PdfPipelineOptions,
    )
    from docling.document_converter import DocumentConverter, PdfFormatOption
    from docling.models.ocr_mac_model import OcrMacOptions
    from docling.models.tesseract_ocr_cli_model import TesseractCliOcrOptions
    from docling.models.tesseract_ocr_model import TesseractOcrOptions
    # from docling_core.types.doc import ImageRefMode, PictureItem, TableItem

    start_time = time.time()
    source = "multi_page_table_test.pdf"  # local path or URL to your PDF.

    pipeline_options = PdfPipelineOptions()
    pipeline_options.do_ocr = True
    pipeline_options.do_table_structure = True
    pipeline_options.table_structure_options.do_cell_matching = True
    pipeline_options.ocr_options.lang = ["es"]
    pipeline_options.accelerator_options = AcceleratorOptions(
        num_threads=4, device=AcceleratorDevice.AUTO
    )

    doc_converter = DocumentConverter(
        format_options={
            InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)
        }
    )

    result = doc_converter.convert(source)
    extraction_time = time.time() - start_time
    print(f"Total extraction time: {extraction_time:.2f} seconds \n\n")
    print(f"docling document= {result.document.export_to_dict()}")

    tables = result.document.tables

    merged_tables = merge_tables_list(
        tables
        # bbox_threshold=10.0,
        # bottom_threshold=100,
        # top_threshold=800,
        # header_sim_threshold=0.7
    )

    print("\nFinal Merged Tables:\n")
    for idx, table in enumerate(merged_tables, start=1):
        merge_ids = table.get('merged_ids', [])
        if len(merge_ids) > 1:
            print(f"Tables {', '.join(merge_ids)} are merged into one:")
        else:
            print(f"Table {merge_ids[0]} remains standalone.")

        prov = table['prov'][0]
        page = prov['page_no']
        grid = table['data'].get('grid', [])
        col_count = get_col_count(table)
        print(f"  Starting Page: {page}")
        print(f"  Number of Columns: {col_count}")
        print("  Rows:")
        for r, row in enumerate(grid, start=1):
            row_text = " | ".join(cell.get('text', '') for cell in row)
            print(f"    Row {r}: {row_text}")
        print()
```

```python
from copy import deepcopy
from sentence_transformers import SentenceTransformer, util
import time

# Initialize the model for text similarity.
model = SentenceTransformer('all-MiniLM-L6-v2')

def to_dict(table_item):
    if isinstance(table_item, dict):
        return table_item
    elif hasattr(table_item, "dict"):
        return table_item.dict()
    return table_item

def bbox_match(bbox1, bbox2, threshold=10.0):
    # Check that key coordinates (left, right, top) are within the threshold.
    left_diff = abs(bbox1['l'] - bbox2['l'])
    right_diff = abs(bbox1['r'] - bbox2['r'])
    top_diff = abs(bbox1['t'] - bbox2['t'])
    return (left_diff <= threshold and right_diff <= threshold and top_diff <= threshold)

def merge_bbox(bbox1, bbox2):
    # Merge two bounding boxes so that the new bbox spans both.
    return {
        'l': min(bbox1['l'], bbox2['l']),
        'r': max(bbox1['r'], bbox2['r']),
        'b': min(bbox1['b'], bbox2['b']),
        't': max(bbox1['t'], bbox2['t']),
        'coord_origin': bbox1.get('coord_origin', 'BOTTOMLEFT')
    }

def get_col_count(table):
    grid = table.get('data', {}).get('grid', [])
    return len(grid[0]) if grid and grid[0] else 0

def extract_first_data_row_text(table):
    """
    Extract the first data row (i.e. second row if a header is present).
    If only one row exists, return that row.
    """
    grid = table.get('data', {}).get('grid', [])
    if len(grid) > 1:
        return " | ".join(cell.get('text', '').strip() for cell in grid[1])
    elif grid:
        return " | ".join(cell.get('text', '').strip() for cell in grid[0])
    return ""

def extract_last_data_row_text(table):
    """
    Extract the last data row from the table.
    If only one row exists, return that row.
    """
    grid = table.get('data', {}).get('grid', [])
    if len(grid) > 1:
        return " | ".join(cell.get('text', '').strip() for cell in grid[-1])
    elif grid:
        return " | ".join(cell.get('text', '').strip() for cell in grid[0])
    return ""

def compute_text_similarity(text1, text2):
    if not text1 or not text2:
        return 0.0
    emb1 = model.encode(text1, convert_to_tensor=True)
    emb2 = model.encode(text2, convert_to_tensor=True)
    return util.pytorch_cos_sim(emb1, emb2).item()

def merge_tables_list(tables, data_row_sim_threshold=0.5, bbox_threshold=10.0):
    """
    Generalized merge:
      - Only tables on consecutive pages are considered.
      - Their bounding boxes (key coordinates) must match within bbox_threshold.
      - The similarity between the last data row of the current table and the first data row of the candidate
        must be at or above data_row_sim_threshold.
      - They must have the same number of columns.
    """
    tables = [to_dict(t) for t in tables]
    merged_tables = []
    i = 0
    while i < len(tables):
        # Start with the current table.
        current_table = deepcopy(tables[i])
        merge_ids = [current_table.get('self_ref', f"Table_{i}")]
        current_page = current_table['prov'][0]['page_no']
        current_bbox = current_table['prov'][0]['bbox']
        current_cols = get_col_count(current_table)
        merged_grid = current_table.get('data', {}).get('grid', [])[:]

        print(f"\nStarting merge with Table {merge_ids[0]} on page {current_page}")
        j = i + 1
        while j < len(tables):
            candidate = tables[j]
            candidate_page = candidate['prov'][0]['page_no']
            # Rule 1: Only merge tables on consecutive pages.
            if candidate_page == current_page:
                break  # Same page—do not merge.
            if candidate_page != current_page + 1:
                break  # Only immediate next page is allowed.

            candidate_bbox = candidate['prov'][0]['bbox']
            # Rule 2: Bounding box check.
            if not bbox_match(current_bbox, candidate_bbox, threshold=bbox_threshold):
                print(f"--> Not merging Table {candidate.get('self_ref', f'Table_{j}')}: bounding box mismatch.")
                break

            # Rule 3: Column count must match.
            candidate_cols = get_col_count(candidate)
            if candidate_cols != current_cols:
                print(f"--> Not merging Table {candidate.get('self_ref', f'Table_{j}')}: column count differs.")
                break

            # Rule 4: Content similarity check on representative data rows.
            current_last = extract_last_data_row_text(current_table)
            candidate_first = extract_first_data_row_text(candidate)
            data_sim = compute_text_similarity(current_last, candidate_first)
            if data_sim < data_row_sim_threshold:
                print(f"--> Not merging Table {candidate.get('self_ref', f'Table_{j}')}: data row similarity too low ({data_sim:.2f}).")
                break

            # All checks passed; merge the candidate.
            print(f"--> Merging Table {candidate.get('self_ref', f'Table_{j}')}: data row similarity = {data_sim:.2f}.")
            candidate_grid = candidate.get('data', {}).get('grid', [])
            # Assume candidate table has a header row; skip it.
            merge_rows = candidate_grid[1:] if len(candidate_grid) > 1 else candidate_grid
            merge_ids.append(candidate.get('self_ref', f"Table_{j}"))
            merged_grid.extend(merge_rows)

            # Update bounding box and current page info.
            current_bbox = merge_bbox(current_bbox, candidate_bbox)
            current_table['prov'][0]['bbox'] = current_bbox
            current_page = candidate_page
            j += 1

        current_table['data']['grid'] = merged_grid
        current_table['merged_ids'] = merge_ids
        merged_tables.append(current_table)
        i = j  # Move to next unmerged table.

    return merged_tables

# --- Example Usage ---
if __name__ == "__main__":
    from docling.document_converter import DocumentConverter
    from docling.backend.pypdfium2_backend import PyPdfiumDocumentBackend
    from docling.datamodel.base_models import InputFormat
    from docling.datamodel.pipeline_options import AcceleratorDevice, AcceleratorOptions, PdfPipelineOptions
    from docling.document_converter import DocumentConverter, PdfFormatOption
    from docling.models.ocr_mac_model import OcrMacOptions
    from docling.models.tesseract_ocr_cli_model import TesseractCliOcrOptions
    from docling.models.tesseract_ocr_model import TesseractOcrOptions

    start_time = time.time()
    source = "merge_logic_negative_scenario_testing.pdf"  # Test PDF
    pipeline_options = PdfPipelineOptions()
    pipeline_options.do_ocr = True
    pipeline_options.do_table_structure = True
    pipeline_options.table_structure_options.do_cell_matching = True
    pipeline_options.ocr_options.lang = ["es"]
    pipeline_options.accelerator_options = AcceleratorOptions(num_threads=4, device=AcceleratorDevice.AUTO)

    doc_converter = DocumentConverter(
        format_options={InputFormat.PDF: PdfFormatOption(pipeline_options=pipeline_options)}
    )

    result = doc_converter.convert(source)
    extraction_time = time.time() - start_time
    print(f"Total extraction time: {extraction_time:.2f} seconds\n")
    print(f"docling document= {result.document.export_to_dict()}")

    tables = result.document.tables
    # Use default thresholds; these can be adjusted globally if needed.
    merged_tables = merge_tables_list(tables, data_row_sim_threshold=0.5, bbox_threshold=10.0)

    print("\nFinal Merged Tables:\n")
    for idx, table in enumerate(merged_tables, start=1):
        merge_ids = table.get('merged_ids', [])
        if len(merge_ids) > 1:
            print(f"Tables {', '.join(merge_ids)} are merged into one:")
        else:
            print(f"Table {merge_ids[0]} remains standalone.")
        prov = table['prov'][0]
        page = prov['page_no']
        grid = table['data'].get('grid', [])
        col_count = get_col_count(table)
        print(f"  Starting Page: {page}")
        print(f"  Number of Columns: {col_count}")
        print("  Rows:")
        for r, row in enumerate(grid, start=1):
            row_text = " | ".join(cell.get('text', '') for cell in row)
            print(f"    Row {r}: {row_text}")
        print()
```
